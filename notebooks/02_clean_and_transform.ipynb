```python
from pathlib import Path
import re
import pandas as pd
import numpy as np

# Required raw Excel inputs (header=1)
table_files = [
    "balancesheet.xlsx",
    "cashflow.xlsx",
    "profitandloss.xlsx",
]

candidate_dirs = [
    Path("data/raw"),
    Path("../data/raw"),
    Path("../../data/raw"),
]

source_files = []
for name in table_files:
    found = next((str(d / name) for d in candidate_dirs if (d / name).exists()), None)
    if found:
        source_files.append(found)

if len(source_files) != len(table_files):
    raise FileNotFoundError(
        f"Expected files: {table_files}. Found: {source_files}. Checked: {[str(d) for d in candidate_dirs]}"
    )

labels = set()
candidate_period_cols = ["date", "year", "period", "fiscal_year", "reporting_period"]

for file in source_files:
    df = pd.read_excel(file, header=1)

    # Wide-format extraction: periods in column headers (skip first descriptor/id column).
    cols = df.columns[1:]
    labels.update(str(c).strip() for c in cols if str(c).strip())

    # Long-format fallback: periods in rows under known period columns.
    normalized_cols = {str(col).strip().lower(): col for col in df.columns}
    period_col = next((normalized_cols[c] for c in candidate_period_cols if c in normalized_cols), None)
    if period_col is not None:
        values = df[period_col].dropna().astype(str).str.strip()
        labels.update(v for v in values if v)

if not labels:
    raise ValueError("No valid period labels found in source files.")

dim_year = pd.DataFrame({"raw_label": sorted(labels)})
dim_year["clean_label"] = dim_year["raw_label"].str.strip()
dim_year["is_ttm"] = dim_year["clean_label"].str.contains(r"\bTTM\b", case=False, na=False)

month_abbr_to_num = {
    "JAN": 1, "FEB": 2, "MAR": 3, "APR": 4,
    "MAY": 5, "JUN": 6, "JUL": 7, "AUG": 8,
    "SEP": 9, "OCT": 10, "NOV": 11, "DEC": 12,
}

def parse_month_year(label: str):
    text = str(label).strip()

    if not text or "TTM" in text.upper():
        return pd.NA, pd.NA, pd.NA

    parsed = pd.to_datetime(text, errors="coerce")
    if pd.notna(parsed):
        return int(parsed.month), int(parsed.year), parsed.strftime("%b")

    # Fallback for labels like "Mar 2024", "Sep-2023", "Dec/2022 9M"
    match = re.search(r"([A-Za-z]{3,9})\s*[-/]?\s*(\d{4})", text)
    if match:
        month_key = match.group(1)[:3].upper()
        year_val = int(match.group(2))
        month_num = month_abbr_to_num.get(month_key, pd.NA)
        if pd.notna(month_num):
            month_str = month_key.title()
            return int(month_num), year_val, month_str

    return pd.NA, pd.NA, pd.NA

parsed_df = dim_year["clean_label"].apply(
    lambda x: pd.Series(parse_month_year(x), index=["month", "year", "month_str"])
 )
dim_year = pd.concat([dim_year, parsed_df], axis=1)

# Keep only valid calendar periods + TTM
valid_period = dim_year["month"].notna() & dim_year["year"].notna()
dim_year = dim_year[dim_year["is_ttm"] | valid_period].copy()

# Remove clearly invalid years (e.g., Mar 1 -> year=1)
year_int = dim_year["year"].astype("Int64")
valid_year_range = year_int.between(1990, 2100, inclusive="both")
dim_year = dim_year[dim_year["is_ttm"] | valid_year_range].copy()

if dim_year.empty:
    raise ValueError("Period parsing produced no valid rows. Check source period labels and parser rules.")

# year_label: e.g., "Mar 2024"
dim_year["year_label"] = np.where(
    dim_year["is_ttm"],
    "TTM",
    dim_year["month_str"] + " " + dim_year["year"].astype("Int64").astype(str)
)

# Fiscal year rule (Apr-Mar)
dim_year["fiscal_year"] = dim_year["year"].astype("Int64")
fiscal_rollover = (~dim_year["is_ttm"]) & (dim_year["month"].astype("Int64") >= 4)
dim_year.loc[fiscal_rollover, "fiscal_year"] = dim_year.loc[fiscal_rollover, "fiscal_year"] + 1
dim_year.loc[dim_year["is_ttm"], "fiscal_year"] = pd.NA

def get_quarter(month):
    if pd.isna(month):
        return None
    if month == 3:
        return "Q4"
    elif month == 6:
        return "Q1"
    elif month == 9:
        return "Q2"
    elif month == 12:
        return "Q3"
    return None

dim_year["quarter"] = dim_year["month"].apply(get_quarter)
dim_year.loc[dim_year["is_ttm"], "quarter"] = "TTM"

# Half-year should align to Sep/Q2 periods
dim_year["is_half_year"] = dim_year["month"].astype("Int64").eq(9).fillna(False).astype(bool)

# year_id (INT)
base_period_id = dim_year["year"].astype("Int64") * 100 + dim_year["month"].astype("Int64")
dim_year["year_id"] = np.where(
    dim_year["is_ttm"],
    999999,
    base_period_id
)
dim_year["year_id"] = dim_year["year_id"].astype("Int64")

dim_year["sort_order"] = np.where(
    dim_year["is_ttm"],
    999999,
    base_period_id
)
dim_year["sort_order"] = dim_year["sort_order"].astype("Int64")

# De-duplicate by surrogate key (year_id)
dim_year_final = dim_year[[
    "year_id",
    "year_label",
    "fiscal_year",
    "quarter",
    "is_ttm",
    "is_half_year",
    "sort_order",
]].sort_values(["sort_order", "year_label"], na_position="last")
dim_year_final = dim_year_final.drop_duplicates(subset=["year_id"], keep="first")

dim_year_final.to_csv("data/clean/dim_year.csv", index=False)
dim_year_final.head()
```

```python
from pathlib import Path
import re
import numpy as np
import pandas as pd

output_dir = Path("data/clean")
output_dir.mkdir(parents=True, exist_ok=True)

dim_company = pd.read_csv(output_dir / "dim_company.csv")
dim_year = pd.read_csv(output_dir / "dim_year.csv")

def normalize_period_label(value):
    text = str(value).strip()
    if not text or text.lower() in {"nan", "none"}:
        return None
    if "TTM" in text.upper():
        return "TTM"
    parsed = pd.to_datetime(text, errors="coerce")
    if pd.notna(parsed):
        return f"{parsed.strftime('%b')} {int(parsed.year)}"
    match = re.search(r"([A-Za-z]{3,9})\s*[-/]?\s*(\d{2,4})", text)
    if match:
        month = match.group(1)[:3].title()
        year = int(match.group(2))
        if year < 100:
            year += 2000
        return f"{month} {year}"
    return text.upper()

dim_year_lookup = (
    dim_year.assign(label_norm=dim_year["year_label"].map(normalize_period_label))
    .dropna(subset=["label_norm"])
    .drop_duplicates(subset=["label_norm"])
    .set_index("label_norm")["year_id"]
    .to_dict()
 )

def map_year_id(value):
    return dim_year_lookup.get(normalize_period_label(value), pd.NA)

def safe_divide(numerator, denominator):
    numerator = pd.to_numeric(numerator, errors="coerce")
    denominator = pd.to_numeric(denominator, errors="coerce")
    result = np.where(denominator.notna() & (denominator != 0), numerator / denominator, np.nan)
    return pd.Series(result)

# fact_balance_sheet
balance = pd.read_excel("data/raw/balancesheet.xlsx", header=1).copy()
balance.columns = balance.columns.str.strip()
balance["symbol"] = balance["company_id"].astype(str).str.strip().str.upper()
balance["year_id"] = balance["year"].map(map_year_id).astype("Int64")
balance = balance.dropna(subset=["year_id"]).copy()
balance["debt_to_equity"] = safe_divide(balance["borrowings"], balance["equity_capital"] + balance["reserves"])
balance["equity_ratio"] = safe_divide(balance["equity_capital"] + balance["reserves"], balance["total_assets"])
balance["shares_outstanding"] = pd.NA
balance["book_value_per_share"] = pd.NA
fact_balance_sheet = balance[[
    "symbol",
    "year_id",
    "equity_capital",
    "reserves",
    "borrowings",
    "other_liabilities",
    "total_liabilities",
    "fixed_assets",
    "cwip",
    "investments",
    "other_asset",
    "total_assets",
    "debt_to_equity",
    "equity_ratio",
    "shares_outstanding",
    "book_value_per_share",
]].copy()
fact_balance_sheet.to_csv(output_dir / "fact_balance_sheet.csv", index=False)

# fact_profit_loss
profit = pd.read_excel("data/raw/profitandloss.xlsx", header=1).copy()
profit.columns = profit.columns.str.strip()
profit = profit.rename(columns={"company_id": "symbol", "opm_percentage": "opm_pct", "tax_percentage": "tax_pct", "dividend_payout": "dividend_payout_pct"})
profit["symbol"] = profit["symbol"].astype(str).str.strip().str.upper()
profit["year_id"] = profit["year"].map(map_year_id).astype("Int64")
profit = profit.dropna(subset=["year_id"]).copy()

profit = profit.merge(
    fact_balance_sheet[["symbol", "year_id", "total_assets"]],
    on=["symbol", "year_id"],
    how="left",
    suffixes=("", "_bs"),
)
profit["net_profit_margin_pct"] = safe_divide(profit["net_profit"], profit["sales"]) * 100
profit["expense_ratio_pct"] = safe_divide(profit["expenses"], profit["sales"]) * 100
profit["interest_coverage"] = safe_divide(profit["operating_profit"], profit["interest"])
profit["asset_turnover"] = safe_divide(profit["sales"], profit["total_assets"])
profit["return_on_assets"] = safe_divide(profit["net_profit"], profit["total_assets"]) * 100
fact_profit_loss = profit[[
    "symbol",
    "year_id",
    "sales",
    "expenses",
    "operating_profit",
    "opm_pct",
    "other_income",
    "interest",
    "depreciation",
    "profit_before_tax",
    "tax_pct",
    "net_profit",
    "eps",
    "dividend_payout_pct",
    "net_profit_margin_pct",
    "expense_ratio_pct",
    "interest_coverage",
    "asset_turnover",
    "return_on_assets",
]].copy()
fact_profit_loss.to_csv(output_dir / "fact_profit_loss.csv", index=False)

# fact_cash_flow
cash = pd.read_excel("data/raw/cashflow.xlsx", header=1).copy()
cash.columns = cash.columns.str.strip()
cash["symbol"] = cash["company_id"].astype(str).str.strip().str.upper()
cash["year_id"] = cash["year"].map(map_year_id).astype("Int64")
cash = cash.dropna(subset=["year_id"]).copy()
cash["free_cash_flow"] = pd.to_numeric(cash["operating_activity"], errors="coerce") + pd.to_numeric(cash["investing_activity"], errors="coerce")
cash = cash.merge(
    fact_profit_loss[["symbol", "year_id", "net_profit"]],
    on=["symbol", "year_id"],
    how="left",
)
cash["cash_conversion_ratio"] = safe_divide(cash["operating_activity"], cash["net_profit"])
fact_cash_flow = cash[[
    "symbol",
    "year_id",
    "operating_activity",
    "investing_activity",
    "financing_activity",
    "net_cash_flow",
    "free_cash_flow",
    "cash_conversion_ratio",
]].copy()
fact_cash_flow.to_csv(output_dir / "fact_cash_flow.csv", index=False)

# fact_analysis
analysis = pd.read_excel("data/raw/analysis.xlsx", header=1).copy()
analysis.columns = analysis.columns.str.strip()
analysis["symbol"] = analysis["company_id"].astype(str).str.strip().str.upper()
target_periods = {"10Y", "5Y", "3Y", "TTM"}
analysis_metric_map = {
    "compounded_sales_growth": "compounded_sales_growth_pct",
    "compounded_profit_growth": "compounded_profit_growth_pct",
    "stock_price_cagr": "stock_price_cagr_pct",
    "roe": "roe_pct",
}

def parse_analysis_cell(text):
    text = str(text).strip()
    match = re.search(r"(?i)(TTM|10\s*Years?|5\s*Years?|3\s*Years?)\s*:?\s*([-+]?\d+(?:\.\d+)?)\s*%", text)
    if not match:
        return None, np.nan
    period_raw = match.group(1).upper().replace(" ", "")
    period_label = {"10YEARS": "10Y", "5YEARS": "5Y", "3YEARS": "3Y", "TTM": "TTM"}.get(period_raw)
    return period_label, float(match.group(2))

analysis_records = []
for _, row in analysis.iterrows():
    per_period = {}
    for source_col, target_col in analysis_metric_map.items():
        period_label, metric_value = parse_analysis_cell(row[source_col])
        if period_label in target_periods:
            per_period.setdefault(period_label, {})[target_col] = metric_value
    for period_label, metrics in per_period.items():
        analysis_records.append({
            "symbol": row["symbol"],
            "period_label": period_label,
            **metrics,
        })

fact_analysis = pd.DataFrame(analysis_records)
if not fact_analysis.empty:
    fact_analysis = fact_analysis.sort_values(["symbol", "period_label"]).drop_duplicates(subset=["symbol", "period_label"], keep="first")
fact_analysis.to_csv(output_dir / "fact_analysis.csv", index=False)

# fact_pros_cons
pros_cons = pd.read_excel("data/raw/prosandcons.xlsx", header=1).copy()
pros_cons.columns = pros_cons.columns.str.strip()
pros_cons["symbol"] = pros_cons["company_id"].astype(str).str.strip().str.upper()
pros_cons["generated_at"] = pd.Timestamp.now(tz="UTC").floor("s")

pros_rows = pros_cons[["symbol", "pros", "generated_at"]].dropna(subset=["pros"]).copy()
pros_rows = pros_rows.rename(columns={"pros": "text"})
pros_rows["is_pro"] = True
pros_rows["category"] = "pros"
pros_rows["source"] = "MANUAL"
pros_rows["confidence"] = 1.0

cons_rows = pros_cons[["symbol", "cons", "generated_at"]].dropna(subset=["cons"]).copy()
cons_rows = cons_rows.rename(columns={"cons": "text"})
cons_rows["is_pro"] = False
cons_rows["category"] = "cons"
cons_rows["source"] = "MANUAL"
cons_rows["confidence"] = 1.0

fact_pros_cons = pd.concat([pros_rows, cons_rows], ignore_index=True)[[
    "symbol",
    "is_pro",
    "category",
    "text",
    "source",
    "confidence",
    "generated_at",
]]
fact_pros_cons.to_csv(output_dir / "fact_pros_cons.csv", index=False)

# fact_ml_scores placeholder until a model output source is available
fact_ml_scores = pd.DataFrame(columns=[
    "symbol",
    "computed_at",
    "overall_score",
    "profitability_score",
    "growth_score",
    "leverage_score",
    "cashflow_score",
    "dividend_score",
    "trend_score",
    "health_label",
])
fact_ml_scores.to_csv(output_dir / "fact_ml_scores.csv", index=False)

print("fact_balance_sheet:", fact_balance_sheet.shape)
print("fact_profit_loss:", fact_profit_loss.shape)
print("fact_cash_flow:", fact_cash_flow.shape)
print("fact_analysis:", fact_analysis.shape)
print("fact_pros_cons:", fact_pros_cons.shape)
print("fact_ml_scores:", fact_ml_scores.shape)
fact_profit_loss.head()
```
